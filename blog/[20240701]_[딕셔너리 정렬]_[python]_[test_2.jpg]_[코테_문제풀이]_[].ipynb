## 성격유형 검사하기

나만의 카카오 성격 유형 검사지를 만들려고 합니다.
성격 유형 검사는 다음과 같은 4개 지표로 성격 유형을 구분합니다. 성격은 각 지표에서 두 유형 중 하나로 결정됩니다.

| 지표 번호 | 성격 유형 |
| -- | -- |
|1번 지표 | 라이언형(R), 튜브형(T) |
|2번 지표 | 콘형(C), 프로도형(F) |
|3번 지표 | 제이지형(J), 무지형(M) |
|4번 지표 | 어피치형(A), 네오형(N) |

4개의 지표가 있으므로 성격 유형은 총 16(=2 x 2 x 2 x 2)가지가 나올 수 있습니다. 예를 들어, "RFMN"이나 "TCMA"와 같은 성격 유형이 있습니다.

검사지에는 총 n개의 질문이 있고, 각 질문에는 아래와 같은 7개의 선택지가 있습니다.

매우 비동의/비동의/약간 비동의/모르겠음/약간 동의/동의/매우 동의

각 질문은 1가지 지표로 성격 유형 점수를 판단합니다.

예를 들어, 어떤 한 질문에서 4번 지표로 아래 표처럼 점수를 매길 수 있습니다.

|선택지 | 성격 유형 점수 |
| -- |--|
|매우 비동의 | 네오형 3점 |
|비동의 | 네오형 2점 |
|약간 비동의 | 네오형 1점 |
|모르겠음 | 어떤 성격 유형도 점수를 얻지 않습니다 |
|약간 동의 | 어피치형 1점 |
|동의 | 어피치형 2점 |
|매우 동의 |어피치형 3점 |

이때 검사자가 질문에서 약간 동의 선택지를 선택할 경우 어피치형(A) 성격 유형 1점을 받게 됩니다. 만약 검사자가 매우 비동의 선택지를 선택할 경우 네오형(N) 성격 유형 3점을 받게 됩니다.

위 예시처럼 네오형이 비동의, 어피치형이 동의인 경우만 주어지지 않고, 질문에 따라 네오형이 동의, 어피치형이 비동의인 경우도 주어질 수 있습니다.
하지만 각 선택지는 고정적인 크기의 점수를 가지고 있습니다.

- 매우 동의나 매우 비동의 선택지를 선택하면 3점을 얻습니다.
- 동의나 비동의 선택지를 선택하면 2점을 얻습니다.
- 약간 동의나 약간 비동의 선택지를 선택하면 1점을 얻습니다.
- 모르겠음 선택지를 선택하면 점수를 얻지 않습니다.

검사 결과는 모든 질문의 성격 유형 점수를 더하여 각 지표에서 더 높은 점수를 받은 성격 유형이 검사자의 성격 유형이라고 판단합니다. 단, 하나의 지표에서 각 성격 유형 점수가 같으면, 두 성격 유형 중 사전 순으로 빠른 성격 유형을 검사자의 성격 유형이라고 판단합니다.

질문마다 판단하는 지표를 담은 1차원 문자열 배열 survey와 검사자가 각 질문마다 선택한 선택지를 담은 1차원 정수 배열 choices가 매개변수로 주어집니다. 이때, 검사자의 성격 유형 검사 결과를 지표 번호 순서대로 return 하도록 solution 함수를 완성해주세요.

제한사항

1 ≤ survey의 길이 ( = n) ≤ 1,000

survey의 원소는 "RT", "TR", "FC", "CF", "MJ", "JM", "AN", "NA" 중 하나입니다.

survey[i]의 첫 번째 캐릭터는 i+1번 질문의 비동의 관련 선택지를 선택하면 받는 성격 유형을 의미합니다.

survey[i]의 두 번째 캐릭터는 i+1번 질문의 동의 관련 선택지를 선택하면 받는 성격 유형을 의미합니다.

choices의 길이 = survey의 길이

choices[i]는 검사자가 선택한 i+1번째 질문의 선택지를 의미합니다.

1 ≤ choices의 원소 ≤ 7

| choices | 뜻 |
| -- | -- |
| 1 | 매우 비동의 |
| 2 | 비동의 |
| 3 | 약간 비동의 |
| 4 | 모르겠음 |
| 5 | 약간 동의 |
| 6 | 동의 |
| 7 | 매우 동의 |

소요시간: 3시간... 문제를 너무 복잡하게 생각해서 오래걸렸다. 있지도 않은 조건을 더 넣느라..

````python
# 1. 성격별 점수 딕셔너리, survey에서 나올 수 있는 경우의 수 리스트 생성

# 2. 입력받은 survey, choices 값으로 딕셔너리 value에 스코어 합산

# 3. 전체 딕셔너리에서 탈락한 성격유형 제거

# 4. 닥셔너리의 키 값만 추출하여 리스트로 정리

# 5. 정리된 리스트를 문자열로 변경(join 함수 사용)

def solution(survey, choices):
    # 1
    dict_mbti = {"R": 0, "T": 0, "F": 0, "C": 0, "J": 0, "M": 0, "A": 0, "N": 0}
    survey_elements = ["RT", "TR", "FC", "CF", "MJ", "JM", "AN", "NA"]

    # 2
    for i in range(len(survey)):
        for j in survey_elements:
            if survey[i] == j:
                if 1 <= choices[i] <= 3:
                    dict_mbti[j[0]] += (4 - choices[i])
                elif 5 <= choices[i] <= 7:
                    dict_mbti[j[1]] += (choices[i] - 4)
                else:
                    pass

    # 3
    if dict_mbti['R'] < dict_mbti['T']:
        dict_mbti.pop('R')
    else:
        dict_mbti.pop('T')

    if dict_mbti['C'] < dict_mbti['F']:
        dict_mbti.pop('C')
    else:
        dict_mbti.pop('F')

    if dict_mbti['J'] < dict_mbti['M']:
        dict_mbti.pop('J')
    else:
        dict_mbti.pop('M')

    if dict_mbti['A'] < dict_mbti['N']:
        dict_mbti.pop('A')
    else:
        dict_mbti.pop('N')

    dict_key = list(dict_mbti.keys())
    result = "".join(dict_key)

    return result
````

### 에러 정리

1. 등호 순서: '=<' 가 아니라 '<=' 가 맞는 표현, 헷갈리지 말 것

이것 말곤 에러는 없었다. 하지만, 문제 조건을 잘못 이해해서 많이 틀리고 헤맸다.

이렇게 간단하게 풀 수 있는걸

나는 **각 성격 유형별로 뽑은 다음, 가장 점수가 높은 순서대로 다시 정렬하고**

**점수가 같다면, 문자열 순서대로 정렬하여 나타내라** 라고 이해했다.

그러다 보니, 정렬도 해야 했고 value 값이 같다면 그 같은 키 값을 다시 정렬해야 했다.

또한, 하나의 딕셔너리에서 숫자 순서대로 한 다음 문자 정렬을 따로 해야 했기 때문에 너무 어려웠다.

다시 생각해보면 mbti는 정해진 순서가 있는데 왜 그렇게 생각했는지 모르겠다...

다음은 초기에 생각했던 대로 한 코드다.

````python
# 1. 성격별 점수 딕셔너리, survey에서 나올 수 있는 경우의 수 리스트 생성

# 2. 입력받은 survey, choices 값으로 딕셔너리 value에 스코어 합산

# 3. 전체 딕셔너리에서 탈락한 성격유형 제거

# 4. 딕셔너리 value 값으로 내림차순 정렬

# 5. value 값이 같은 키끼리 키를 기준으로 오름차순 정렬

# 6. 키 값을 추출하여 문자열 형태로 변환

def solution(survey, choices):
    # 1
    dict_mbti = {"R": 0, "T": 0, "F": 0, "C": 0, "J": 0, "M": 0, "A": 0, "N": 0}
    survey_elements = ["RT", "TR", "FC", "CF", "MJ", "JM", "AN", "NA"]

    # 2
    for i in range(len(survey)):
        for j in survey_elements:
            if survey[i] == j:
                if 1 <= choices[i] <= 3:
                    dict_mbti[j[0]] += (4 - choices[i])
                elif 5 <= choices[i] <= 7:
                    dict_mbti[j[1]] += (choices[i] - 4)
                else:
                    pass

    # 3
    if dict_mbti['R'] < dict_mbti['T']:
        dict_mbti.pop('R')
    else:
        dict_mbti.pop('T')

    if dict_mbti['C'] < dict_mbti['F']:
        dict_mbti.pop('C')
    else:
        dict_mbti.pop('F')

    if dict_mbti['J'] < dict_mbti['M']:
        dict_mbti.pop('J')
    else:
        dict_mbti.pop('M')

    if dict_mbti['A'] < dict_mbti['N']:
        dict_mbti.pop('A')
    else:
        dict_mbti.pop('N')

    # 4
    sorted_dict1 = dict(sorted(dict_mbti.items(), key= lambda item:item[1], reverse=True))

    # 5.1 특정 키, value 가 같은 것만 따로 딕셔너리 생성
    sorted_dict2 = {}
    for key1, value1 in sorted_dict1.items():
        for key2, value2 in sorted_dict1.items():
            if key1 != key2 and sorted_dict1[key1] == sorted_dict1[key2]:
                sorted_dict2[key1] = value1
                sorted_dict2[key2] = value2

    sorted_dict2 = dict(sorted(sorted_dict2.items(), key= lambda item:item[0]))

    # 5.2 다음으로 두 딕셔너리 합치기(순서도 업데이트해야 함)
    sorted_dict2.update(sorted_dict1)
    sorted_dict3 = dict(sorted(sorted_dict2.items(), key= lambda item:item[1], reverse=True))

    # 6
    result = list(sorted_dict3.keys())
    result = "".join(result)

    return result
````

4번 부분부터 막혔다.

딕셔너리의 value 값을 토대로 내림차순 정렬을 해야 했다.

내림차순 정렬을 하는 방식은 다음과 같다.

sorted()로 정렬을 하겠다는 것을 알려주고, items()로 리스트 안에 튜플 형식으로 바꾼다. 

key = lambda item: item[1], reverse = True, 아이템의 뒤의 값(value)를 reverse(뒤집다.)를 true로 하여 내림차순으로 정렬한다.

오름차순의 경우, reverse = False로 한다.

정리하면, 

```py
sorted_dict1 = dict(sorted(dict_mbti.items(), key= lambda item:item[1], reverse=True))
```
이고 sorted를 하는 경우, 리스트 형식이 되기 때문에 이를 다시 딕셔너리로 바꿔준 다음 재정의한다.

그러면 문자열에 관계없이 받아온 딕셔너리에서 value 값 순서대로 정렬된 딕셔너리가 완성된다.

여기서 value 값이 같은 경우, 문자열로 정렬해야 하는데

그냥 문자열 정렬을 하는 경우, 앞서 value 값으로 정렬한 것이 초기화되기 때문에

따로 똑같은 value 값을 가지는 items를 뽑아 빈 딕셔너리에 넣어줘야 한다.

그 코드는 다음과 같다.

```py
sorted_dict2 = {}
for key1, value1 in sorted_dict1.items():
    for key2, value2 in sorted_dict1.items():
        if key1 != key2 and sorted_dict1[key1] == sorted_dict1[key2]:
            sorted_dict2[key1] = value1
            sorted_dict2[key2] = value2
```

중첩 for 문을 통해 같은 key는 다르지만 value는 같은 딕셔너리 값을 sorted_dict2에 넣어준다.

이를 이제 key 값으로 오름차순 정리하면 된다.

앞서 언급했듯, reverse = False를 넣어도 되지만, 오름차순이 defualt 값이기 때문에 생략 가능하다.

```py
sorted_dict2 = dict(sorted(sorted_dict2.items(), key= lambda item:item[0]))
```

이번에는 키 값으로 정렬이기 때문에 item[0]으로 쓴다.

**두 딕셔너리를 합치는 방법**은 다음과 같다.

sorted_dict2.update(sorted_dict1)

동일한 키 값을 갖고 있다면 앞 딕셔너리의 순서를 따르기 때문에 sorted_dict2인 새로 만든 딕셔너리에 기존 딕셔너리 sorted_dict1을 합친다.

이러면, 또 value 값 정렬이 안되기 때문에 다시 value 값 정렬을 해준다.

마지막으로 이를 sorted_dict3.key()를 통해 key 값을 불러오고 "".join()을 통해 문자열로 바꿔 주면 값이 나온다.

뻘짓을 통해 새로 배운 내용은 다음과 같다.

**딕셔너리 정렬**

key 값으로 오름차순 정렬하는 경우,

```py
sorted(sorted_dict2.items(), key= lambda item:item[0])
```

key 값으로 내림차순 정렬하는 경우,

```py
sorted(sorted_dict2.items(), key= lambda item:item[0], reverse = True)
```

value 값으로 오름차순 정렬하는 경우, item[1]로 바꿔주면 된다.

**리스트를 문자열로**

```py
"".join()
```

**딕셔너리 업데이트**

```py
dict1.update(dict2) # main 딕셔너리를 앞에!
```

다음은 위 문제를 ai가 풀어준 결과다.

````python
def solution(survey, choices):
    # 1. 성격별 점수 딕셔너리 초기화
    scores = {
        'R': 0, 'T': 0,
        'C': 0, 'F': 0,
        'J': 0, 'M': 0,
        'A': 0, 'N': 0
    }
    
    # 2. 입력받은 survey와 choices를 기반으로 점수 합산
    for s, choice in zip(survey, choices):
        if choice < 4:
            scores[s[0]] += 4 - choice
        elif choice > 4:
            scores[s[1]] += choice - 4
    
    # 3. 각 지표에서 더 높은 점수를 받은 성격 유형 결정
    result = []
    result.append('R' if scores['R'] >= scores['T'] else 'T')
    result.append('C' if scores['C'] >= scores['F'] else 'F')
    result.append('J' if scores['J'] >= scores['M'] else 'M')
    result.append('A' if scores['A'] >= scores['N'] else 'N')
    
    # 4. 결과를 문자열로 변환하여 반환
    return ''.join(result)
````
